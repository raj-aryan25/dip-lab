WAP to appy transformations on an image (scaling, rotation, resizing)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Load image
img = cv2.imread("./content/cat.jpg")
if img is None:
    raise FileNotFoundError("cat.jpg not found")
```

```python
# Display original image
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title("Original")
plt.show()
```

```python
# Scaling (uniform)
scale_factor = 1.5
scaled = cv2.resize(img, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_LINEAR)
scaled_rgb = cv2.cvtColor(scaled, cv2.COLOR_BGR2RGB)
plt.imshow(scaled_rgb)
plt.title("Scaled")
plt.show()
```

```python
# Rotation (around center)
(h, w) = img.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img, M, (w, h))
rotated_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)
plt.imshow(rotated_rgb)
plt.title("Rotated")
plt.axis("off")
plt.show()
```

```python
# Resizing to fixed size
resized = cv2.resize(img, (300, 300), interpolation=cv2.INTER_AREA)
resized_rgb = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)
plt.imshow(resized_rgb)
plt.title("Resized to 300x300")
plt.show()
```

```python
# Negating (invert colors)
negated = cv2.bitwise_not(img)
negated_rgb = cv2.cvtColor(negated, cv2.COLOR_BGR2RGB)
plt.imshow(negated_rgb)
plt.title("Negated")
plt.show()
```

```python
# Save results
cv2.imwrite("cat_scaled.jpg", scaled)
cv2.imwrite("cat_rotated.jpg", rotated)
cv2.imwrite("cat_resized.jpg", resized)
cv2.imwrite("cat_negated.jpg", negated)
```
