WAP to apply transformations on an image (scaling, rotation, resizing) without using cv2

```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Load image
img = Image.open("./content/cat.jpg")
img_rgb = img.convert("RGB")
```

```python
# Display original image
plt.imshow(img_rgb)
plt.title("Original")
plt.show()
```

```python
# Scaling (uniform)
scale_factor = 1.5
new_size = (int(img_rgb.width * scale_factor), int(img_rgb.height * scale_factor))
scaled = img_rgb.resize(new_size, Image.BILINEAR)
plt.imshow(scaled)
plt.title("Scaled")
plt.show()
```

```python
# Rotation (around center)
rotated = img_rgb.rotate(45, resample=Image.BILINEAR, expand=False)
plt.imshow(rotated)
plt.title("Rotated")
plt.axis("off")
plt.show()
```

```python
# Resizing to fixed size
resized = img_rgb.resize((300, 300), Image.LANCZOS)
plt.imshow(resized)
plt.title("Resized to 300x300")
plt.show()
```

```python
# Negating (invert colors)
negated = Image.fromarray(255 - np.array(img_rgb))
plt.imshow(negated)
plt.title("Negated")
plt.axis("off")
plt.show()
```

```python
# Save results
scaled.save("cat_scaled.jpg")
rotated.save("cat_rotated.jpg")
resized.save("cat_resized.jpg")
negated.save("cat_negated.jpg")
```
